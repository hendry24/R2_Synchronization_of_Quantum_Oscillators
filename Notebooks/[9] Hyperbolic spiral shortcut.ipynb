```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
import pyqosc as qo
import scipy as sp
```

```python
N = 20
vdp = qo.vdp(N=N, omega = 1, gamma_1 = 2, gamma_2 = 0.1)
rho_0 = qt.coherent_dm(vdp.N, 15+20j)
tau = 2

vdp.shortcut(rho_0, tau, trajectory_func = 0, err_tol_b = 1e-3, timepoints = 101, special = "linear_impens",
                 maxiter = 100, report = True, plot_resulting_trajectory = False, save_to_osc = False,
                 dy = 0)
```
