```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
import pyqosc as qo
```

```python
N = 40
qo.vdp(N = N, 
       omega =3,
       Omega_2 = 1,
       gamma_2 = 1).evolve(rho_0 = qt.coherent_dm(N, 1 * np.exp(1j*np.pi)),
                             timelst = np.linspace(0, 20, 501),
                             expect = [qt.destroy(N)],
                             plot = True)
plt.gca().legend(loc="best")
plt.show()
       
x=1
```
