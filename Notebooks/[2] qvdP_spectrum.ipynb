```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
import pyqosc as qo
import scipy as sp

options = qt.Options(nsteps = int(1e9))
plt.rcParams.update({"font.size" : 13})
```

```python
def sim_1(params):

    N = params.N
    lindblad = qo.vdp_lindblad(params)
    
    omega = np.linspace(-2, 2, 101)
    omega_lim = np.max(omega)
    
    tlst_ss, r, phi, beta_ss = qo.vdp_expvalb(params, t_end = 1e3, t_eval = 1e2, timepoints_returned=int(44100), init_polar=[2, 0])

    omega_q, spect_q, spect_max_q = qo.ss_q_spectrum(lindblad, omega, plot = True)
    omega, spect, spect_max = qo.ss_c_spectrum(tlst_ss, beta_ss, omega_lim, plot = True, overlap_with = plt.gca(), plot_bar = True, color = 'r', label = "cl", width = 1e-2)

    print(f"w_obs (qm) = {spect_max_q} \nw_obs (cl) = {spect_max}")
    plt.show()
```

```python
params = qo.vdp_params(N = 20, Delta = 0.7)
sim_1(params)
```

```python
params = qo.vdp_params(N = 20, Delta = 0, gamma_2 = 3)
sim_1(params)
```
