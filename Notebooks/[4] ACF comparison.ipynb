```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
import pyqosc as qo
import scipy as sp

options = qt.Options(nsteps = int(1e9))
plt.rcParams.update({"font.size" : 13})
```

```python
def sim_1(vdp):

    tlst_ss, r_cycle, phi_cycle, beta_cycle = vdp.adler(t_end = 1e3, 
                                                        t_eval = 9e2,
                                                        timepoints_returned = int(1e4),
                                                        plot = True,
                                                        one_cycle = False)
    ax = plt.gca()
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    qo.ss_c_acf(tlst_ss, beta_cycle, plot = True)
    fig = plt.gcf()
    fig.set_figwidth(10)
    omega, spect, omega_max = qo.ss_c_spectrum(tlst_ss, beta_cycle, omega_lim = 1, plot = True)
    print(omega_max)
    plt.show()
```

```python
vdp = qo.vdp(N = 20, omega = 1.9)
sim_1(vdp)
```

```python
vdp = qo.vdp(N = 20, omega = 1.7, Omega = 0)
sim_1(vdp)
```
